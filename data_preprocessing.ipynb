```python
import pandas as pd
import numpy as np
import data_loader  # å‘¼å«å‰›å‰›çš„æ¬é‹å·¥
```

```python
def clean_data(df):
    print("ğŸ§¹ [Preprocessing] é–‹å§‹æ¸…æ´—è³‡æ–™...")
    df = df.copy()

    # 1. ç¢ºä¿æ—¥æœŸæ’åº
    if 'date' in df.columns:
        df = df.sort_values('date').reset_index(drop=True)

    # 2. è™•ç†ç„¡é™å¤§ (inf)
    # å…ˆæª¢æŸ¥æœ‰æ²’æœ‰ inf
    num_inf = np.isinf(df.select_dtypes(include=np.number)).sum().sum()
    print(f"   - ç™¼ç¾ {num_inf} å€‹ç„¡é™å¤§ (inf) å€¼ï¼Œå°‡è½‰ç‚º NaN...")
    df = df.replace([np.inf, -np.inf], np.nan)

    # 3. è£œç¼ºå€¼
    # å…ˆæª¢æŸ¥æœ‰å¤šå°‘ NaN
    num_nan_before = df.isnull().sum().sum()
    print(f"   - æ¸…æ´—å‰ NaN ç¸½æ•¸: {num_nan_before}")
    
    # åŸ·è¡Œè£œå€¼
    df = df.ffill().fillna(0)
    
    # æª¢æŸ¥è£œå®Œå¾Œé‚„æœ‰æ²’æœ‰
    num_nan_after = df.isnull().sum().sum()
    print(f"   - æ¸…æ´—å¾Œ NaN ç¸½æ•¸: {num_nan_after}")

    return df
```

```python
print("=== æ­¥é©Ÿ 1: è¼‰å…¥åŸå§‹è³‡æ–™ ===")
df_raw = data_loader.load_and_merge_data()

# 2. æ•…æ„æç ´å£ (é¸ç”¨ï¼šæ¸¬è©¦çœ‹çœ‹å¦‚æœæœ‰ NaN æœƒæ€æ¨£)
# df_raw.iloc[0:5, 1] = np.nan # æŠŠå‰äº”å¤©çš„åƒ¹æ ¼è®Šä¸è¦‹
# df_raw.iloc[10, 2] = np.inf  # æŠŠæŸå€‹å€¼è®Šç„¡é™å¤§

# 3. åŸ·è¡Œæ¸…æ´—
print("\n=== æ­¥é©Ÿ 2: åŸ·è¡Œæ¸…æ´— ===")
df_clean = clean_data(df_raw)

# 4. æœ€çµ‚æª¢æŸ¥
print("\n=== æ­¥é©Ÿ 3: æœ€çµ‚æª¢æŸ¥ ===")
print("è³‡æ–™å¤§å°:", df_clean.shape)
print("å‰ 5 ç­†è³‡æ–™:")
print(df_clean.head())

# ç¢ºä¿çœŸçš„æ²’æœ‰ NaN äº† (æ‡‰è©²è¦å°å‡º False)
print("é‚„æœ‰æ®˜ç•™çš„ç©ºå€¼å—ï¼Ÿ", df_clean.isnull().values.any())
```
